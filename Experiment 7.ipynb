```python
import socket

s=socket.socket()

print('Server created')

s.bind(('localhost',20002))

s.listen(3)

print('waiting for connection')
c1, ad=s.accept()

print(ad)
recvdmsg=c1.recv(1024).decode()
print(recvdmsg)
a=int(recvdmsg)+10

c1.send(bytes(str(a).encode()))

c1.close()
s.close()
```

```python
import socket

c=socket.socket()

c.connect(('localhost', 20002))
c.send(bytes('123'.encode()))
recvdmsg1=c.recv(1024).decode()
print(recvdmsg1)

c.close()
```
