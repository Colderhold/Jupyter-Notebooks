**B) Implement Pickle module to store structured data in binary files using dump() and load() methods**

```python
import pickle
class Employee:
  count=0
  def __init__(self, name, id, desig, salary):
    self.name=name
    self.id=id
    self.desig=desig
    self.salary=salary
    Employee.count+=1
  def displayCount(self):
    print("There are %d employees" % Employee.count)
  def displayDetails(self):
    print("Name:", self.name, ", ID:", self.id,", Designation:", self.desig, ", Salary:", self.salary)
e1=Employee("John", "121A0199", "Manager", 80000)
e2=Employee("Mike", "121A0200", "Team Leader", 50000)
e3=Employee("Derek", "121A0201", "Programmer", 30000)
e4=Employee("Raj", "121A0202", "Assistant", 25000)
e4.displayCount()
e1.displayDetails()
e2.displayDetails()
e3.displayDetails()
e4.displayDetails()

# writing on the file
f = 'my_employee.pkl'
ob = open(f,'wb')
pickle.dump(Employee,ob)
ob.close() 

# reading the file
import pickle
f = 'my_employee.pkl'
ob = open(f,'rb')
dis = pickle.load(ob)
print(dis)
ob.close()
```
