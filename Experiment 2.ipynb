**B. Write a Program to Implement Inheritance between classes.**

```python
class teacher:
  def __init__(self,id,name,address):
    self.id=id
    self.name=name
    self.address=address
  
  def display(self):
    print('ID: ',self.id)
    print('NAME: ',self.name)
    print('ADDRESS: ',self.address)

class student(teacher):
  def __init__(self,id,name,address,mark):
    super().__init__(id,name,address)
    self.mark=mark

  def display(self):
    super().display()
    print('Marks scored: ',self.mark)

i=input('Enter the id ')
n=input('Enter the name ')
a=input('Enter the address ')
m=int(input('Enter the marks '))

s=student(i,n,a,m)
s.display()
```

**C. Write a program to show Exception Handling.**

```python
try:
  p=[2,4,6,8]
  print(p[5])
except IndexError:
  print('Index Out of Bounds')
finally:
  print('Final Display')
```

```python
try:
  a=b=c=4
  a=a/(b-c)
except ZeroDivisionError:
  print('Division by 0 Error')
finally:
  print('Final Display')
```
