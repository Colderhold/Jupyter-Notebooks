```python
import numpy as np
import pandas as pd

df = pd.read_csv('Salary_Data.csv')
print(df)
```

```python
x = df[df.columns[0]]
y = df[df.columns[1]]
x_mean = np.mean(x)
y_mean = np.mean(y)
num = np.sum((x - x_mean) * (y - y_mean))
deno = np.sum((x - x_mean) ** 2)
b1 = num / deno
b0 = y_mean - abs(b1 * x_mean)

year = int(input(print('Enter year: ')))
Y = (b1 * year) + b0
print(Y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y, color='green', label='Actual')
plt.plot(x, b1 * x + b0, color='red', label='Predicted')
plt.scatter(x, y, color='green')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary Estimation using Linear Regression')
plt.legend()
plt.show()
```
