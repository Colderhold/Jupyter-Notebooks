```python
user_input = input() 
digits = 0
letters = 0
sum=0 
def isdup(input):
  x=set()
  for i in input:
    if i is not x and input.count(i)>1:
      x.add(i)
  return x
for i in user_input:
 if i.isdigit():
     digits=digits+1
     sum=sum+int(i) 
 elif i.isalpha():
     letters=letters+1
x=isdup(user_input)
print("Duplicates are: ",x)
print(sum) 
print(" The input string",user_input, "has", letters, "letters and", digits,"digits.")
```
