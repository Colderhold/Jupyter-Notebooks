**A. Write a program to take a string from user and check whether it is palindrome or not.**

```python
string=input(("Enter a string:"))
if(string==string[::-1]):
      print("The string is a palindrome")
else:
      print("Not a palindrome")
```


**B. Menu Driven Program:**
1. To create a list and insert an element.
2. To create a tuple and display its last element.
3. To create a set and add one element into it.
4. To create a dictionary and display one of its element from it.

```python
def create_list():
  l = [1,2,'Hello',3]
  print('Before Insertion\n',l)
  l.insert(1,4)
  print('After Insertion\n',l)

def create_tuple():
  t=(7,8,6,'What')
  print('Tuple is\n',t)
  print('Last Element is\n',t[3])

def create_set():
  s={5,4,3,'Are'}
  print('Before Insertion\n',s)
  s.add(9)
  print('After Insertion\n',s)

def create_dictionary():
  d={1:'You',2: 'Doing',3: 'This',4: 'Afternoon',5: 10}
  print('Dictionary is\n',d)
  print("First Element in dictionary is\n",d[1])

n=0
while n!= 5:
  n=int(input(print('\nEnter your choice \n 1. Create List 2. Create Tuple 3. Create Set 4. Create Dictionary')))
  print('Enter 5 to exit\n')
  if n==1:
   create_list()
  elif n==2:
   create_tuple()
  elif n==3:
   create_set()
  elif n==4:
   create_dictionary()
```
