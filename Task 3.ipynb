```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Startups1.csv")
df.head()
```

```python
print(df.shape)
print(df.describe())
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['No. of Employees'], df['Funding Amount in $'])
plt.title('Relationship between No. of Employees and Funding Amount')
plt.xlabel('No. of Employees')
plt.ylabel('Funding Amount in $')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['No. of Investors'], df['Funding Amount in $'])
plt.title('Relationship between No. of Investors and Funding Amount')
plt.xlabel('No. of Investors')
plt.ylabel('Funding Amount in $')
plt.show()
```

```python
grouped_data = df.groupby('Starting Year')['Funding Amount in $'].sum()
plt.bar(grouped_data.index, grouped_data.values)
plt.xlabel('Year')
plt.ylabel('Total Funding Amount in $')
plt.title('Trend of Funding Amounts over Time')
plt.show()
```

```python
plt.scatter(df['Funding Round'], df['Funding Amount in $'])
plt.xlabel('Funding Round')
plt.ylabel('Funding Amount in $')
plt.title('Relationship between Funding Round and Funding Amount')
plt.show()
```

```python
grouped_data = df.groupby('Funding Round')['No. of Investors'].mean()
plt.plot(grouped_data.index, grouped_data.values)
plt.xlabel('Funding Round')
plt.ylabel('Mean Number of Investors')
plt.title('Relationship between Funding Round and Number of Investors')
plt.show()
```
